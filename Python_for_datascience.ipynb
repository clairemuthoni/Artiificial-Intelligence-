### Absolute Deviation and Absolute Mean Deviation using NumPy
Deviation is a measure of the difference between the observed value of a variable and some other value, often that variable’s mean.

Absolute Deviation:

The absolute deviation of an element of a data set is the absolute difference between that element and a given point. The absolute deviation of the observations X1, X2, X3, ….., Xn around a value A is defined as – For discrete (ungrouped) data-

The absolute mean deviation measures the spread and scatteredness of data around, preferably the median value, in terms of absolute deviation. The absolute deviation of observation X1, X2, X3, ……, Xn is minimum when measured around median i.e. A is the median of the data. Then, thus obtained absolute deviation is termed as the absolute mean deviation and is defined as: For discrete (ungrouped) data



```python
# Importing mean, absolute from numpy
from numpy import mean, absolute

data = [75, 69, 56, 46, 47, 79, 92, 97, 89, 88,
        36, 96, 105, 32, 116, 101, 79, 93, 91, 112]

# Assume any point A about which
# absolute deviation is to be calculated
A = 79

sum = 0  # Initialize sum to 0

# Absolute deviation calculation

for i in range(len(data)):
    av = absolute(data[i] - A)   # Absolute value of the differences
                                 # of each data point and A

    # Summing all those absolute values
    sum = sum + av

# Sum divided by length of data yields
# the absolute deviation
print(sum / len(data))
```

```python
# Getting the absolute mean deviation
mean(absolute(data - mean(data)))
```

### Data Cleaning

#### Removal all unwanted observations
this includes deleting duplicate / redundant or irrelevant values from your dataset.

```python
df1 = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
df1.head()
```

#### Handling Missing Data

```python
# Let’s check the % missing values columns-wise for each row using df.isnull()
#it checks whether the values are null or not and gives returns boolean values.
#and .sum() will sum the total number of null values rows and
#we divide it by the total number of rows present in the dataset then we multiply to get values in % i.e per 100 values how much values are null.
round(df1.isna().sum()/len(df1)*100, 2)
```

```python
# Dropping Observations with missing values df1.dropna()
df1.dropna(subset=['Embarked'],axis=0,inplace=True)
df1.head()
```

```python
# we can use mean imputation or median imputation for the case of missing values
# mean imputation is suitable when the data is normally distributed and has no extreme outliers
# Median imputation is preferable when the data contains outliers or is skewed.
df2 = df1.fillna(df1.Age.mean())
df2.isnull().sum()
```

#### Handing outliers
outliers are extreme values that deviate significantly from the majority of the data. They can negatively impact the analysis and model performance.

```python
# To check outliers we use a box plot which is a graphical representation of the datasets distribution
import matplotlib.pyplot as plt

plt.boxplot(df2['Age'], vert=False)
plt.ylabel("Variable")
plt.title("Box Plot")
plt.xlabel('Age')
plt.show()
```

```python
# STEP 1 : Calculate the summary statistics
mean = df2['Age'].mean()
std = df2['Age'].std()
print(mean, std)
```

```python
# STEP 2 : Calculate lower and upper bounds
lower_bound = mean - 3 * std
upper_bound = mean + 3 * std
print(lower_bound, upper_bound)
```

```python
# STEP 3 : Drop the outliers
df3 = df2[(df2['Age'] >= lower_bound) & (df2['Age'] <= upper_bound)]
```

###  Slicing Pandas Dataframe

```python
# Initializing the nested list with Data set
player_list = [['M.S.Dhoni', 36, 75, 5428000],
               ['A.B.D Villers', 38, 74, 3428000],
               ['V.Kohli', 31, 70, 8428000],
               ['S.Smith', 34, 80, 4428000],
               ['C.Gayle', 40, 100, 4528000],
               ['J.Root', 33, 72, 7028000],
               ['K.Peterson', 42, 85, 2528000]]

# creating a pandas dataframe
df = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'])
df # data frame before slicing
```

```python
# Slicing rows in a dataframe in python
df1 = df.iloc[0:4]
df1
```

```python
# Slicing columns in dataframe in python
df2 = df.iloc[:, 0:2]
df2
```

```python
# Selecting a specific cell in dataframe
specific_cell_value = df.iloc[2, 3]  # Row 2, Column 3 (Salary)
print("Specific Cell Value:", specific_cell_value)
```

```python
# Using boolean conditions in dataframe in python
filtered_data = df[df['Age'] > 35].iloc[:, :]  # Select rows where Age is greater than 35
print("\nFiltered Data based on Age > 35:\n", filtered_data)
```
